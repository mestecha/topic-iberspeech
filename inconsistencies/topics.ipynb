```python
def join_turns(data):
    data['CONV_ID'] = data['UID'].str[:-4]
    conversation_ids = data['CONV_ID'].unique()
    sentences = []
    conversations = []
    for conv_id in conversation_ids:
        df_conv = data[data['CONV_ID'] == conv_id]
        df_conv = df_conv.reset_index(drop=True)
        total_turns = df_conv.shape[0]
        for turn in range(0, total_turns-1):
            if df_conv['SEG'].loc[turn][0] == " ":
                sequence = df_conv['SEG'].loc[turn][1:] + df_conv['SEG'].loc[turn+1]
            else:
                sequence = df_conv['SEG'].loc[turn] + df_conv['SEG'].loc[turn+1]

            sentences.append(sequence) 
            conversations.append(conv_id)

    df = pd.DataFrame()
    df['conv_id'] = conversations
    df['sentence'] = sentences
    logging.info(f'Number of registries of resulting df: {df.shape[0]}')
    return df
```

```python
logging.info('Starting...')

logging.info("Loading ZSL pipeline...")
deberta_classifier = pipeline("zero-shot-classification", model = "cross-encoder/nli-deberta-base")

logging.info('Loading Data')
data_path = 'DAILYD_main.csv'
data = pd.read_csv(data_path)
#data = data.loc[0:50000]
data = join_turns(data)

results = predict_topic_improving_DAILY_DIALOG(data, deberta_classifier)
#guardar_datos(results, '/home/jorgemira/nlp/Conversation_Records/Daily-Dialog-Zero-Shot.pkl')
results.to_csv('results.csv', index=False)
```
